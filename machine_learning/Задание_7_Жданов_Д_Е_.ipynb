# Демонстрационный пример

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
X = np.array([2, 3, 5, 7, 8]).reshape(-1, 1)
y = np.array([2, 3, 3, 4, 5])

model = LinearRegression()
model.fit(X, y)
beta_1 = model.coef_[0]
beta_0 = model.intercept_
print(f"Уравнение регрессии: y = {beta_1:.3f}x + {beta_0:.3f}")

hours_6 = np.array([[6]])
prediction_6 = model.predict(hours_6)
print(f"Предсказание для 6 часов учебы: {prediction_6[0]:.3f}")

plt.scatter(X, y, color='blue', label='Фактические данные')
plt.plot(X, model.predict(X), color='red', label='Линия регрессии')
plt.scatter(hours_6, prediction_6, color='green', label='Предсказание (6 часов)')

plt.xlabel('Часы учебы')
plt.ylabel('Оценки')

plt.title('Линейная регрессия: Часы учебы vs Оценки')
plt.legend()
plt.grid(True)
plt.show()
```

# Задание 1. Самостоятельный пример.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = np.array([[5, 80, 75],
              [4, 70, 70],
              [6, 90, 80],
              [3, 60, 65],
              [5.5, 85, 78]])

y = np.array([85, 80, 90, 70, 88])  # итоговые оценки

model = LinearRegression()
model.fit(X, y)

# Коэффициенты для всех трех признаков
beta_1, beta_2, beta_3 = model.coef_
beta_0 = model.intercept_
print(f"Уравнение регрессии: y = {beta_1:.3f}x1 + {beta_2:.3f}x2 + {beta_3:.3f}x3 + {beta_0:.3f}")

# Предсказание
predict_values = np.array([[4.5, 75, 72]])
predict_values_result = model.predict(predict_values)
print(f"Предсказание итоговой оценки: {predict_values_result[0]:.2f}")

y_pred = model.predict(X)

# Часы ДЗ vs Оценка
plt.scatter(X[:, 0], y, color='blue', label='Фактические данные')
plt.plot(X[:, 0], model.predict(X), color='red', label='Линия регрессии', alpha=0.7)
plt.scatter(predict_values[0, 0], predict_values_result, color='green', s=100, label='Предсказание')
plt.legend()
plt.grid(True)
plt.show()

# Дополнительная информация
print(f"\nR² score: {model.score(X, y):.3f}")
print(f"MSE score: {mean_squared_error(y, y_pred):.3f}")
```

# Задание 2. Самостоятельный пример.

```python
X = np.array([[3, 5, 20],
              [5, 7, 25],
              [2, 4, 18],
              [4, 6, 22],
              [6, 8, 30]])

y = np.array([12, 10, 15, 11, 8]) # итоговые оценки

model = LinearRegression()
model.fit(X, y)

# Коэффициенты для всех трех признаков
beta_1, beta_2, beta_3 = model.coef_
beta_0 = model.intercept_
print(f"Уравнение регрессии: y = {beta_1:.3f}x1 + {beta_2:.3f}x2 + {beta_3:.3f}x3 + {beta_0:.3f}")

# Предсказание
predict_values = np.array([[4, 6, 24]])
predict_values_result = model.predict(predict_values)
print(f"Предсказание итоговой оценки: {predict_values_result[0]:.2f}")

y_pred = model.predict(X)

plt.scatter(X[:, 0], y, color='blue', label='Фактические данные')
plt.plot(X[:, 0], model.predict(X), color='red', label='Линия регрессии', alpha=0.7)
plt.scatter(predict_values[0, 0], predict_values_result, color='green', s=100, label='Предсказание')
plt.legend()
plt.grid(True)
plt.show()

# Дополнительная информация
print(f"R² score: {model.score(X, y):.3f}")
print(f"MSE score: {mean_squared_error(y, y_pred):.3f}")

# Анализ важности
feature_names = ['Начальный уровень', 'Часы обучения в неделю', 'Возраст']
abs_coefficients = np.abs(model.coef_)
feature_importance = dict(zip(feature_names, abs_coefficients))

print("\nАбсолютные значения коэффициентов (чем больше, тем важнее):")
for name, importance in sorted(feature_importance.items(), key=lambda x: x[1], reverse=True):
    print(f"{name}: {importance:.4f}")
```
