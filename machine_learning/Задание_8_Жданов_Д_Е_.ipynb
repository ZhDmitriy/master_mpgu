```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = np.array([
    [85, 90], [70, 75], [80, 85], [65, 70], [90, 95],
    [75, 80], [60, 65], [95, 100], [85, 90], [70, 75]
])

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(10, 6))
for i in range(2):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Кластер {i + 1}')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200,
           linewidths=3, color='r', label='Центроиды')

plt.title('Кластеризация студентов по результатам экзаменов')
plt.xlabel('Баллы по математике')
plt.ylabel('Баллы по физике')
plt.grid(True)
plt.legend()
plt.show()

print("Центроиды кластеров:")
for i, centroid in enumerate(centroids):
    print(f"Кластер {i + 1}: {centroid}")
```

# Задание 2. Демонстрационный пример.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Данные: Математика, Физика, Информатика
X = np.array([
    [85, 90, 88], [70, 75, 72], [80, 85, 82], [65, 70, 68], [90, 95, 92],
    [75, 80, 78], [60, 65, 62], [95, 100, 97], [85, 90, 88], [70, 75, 72],
    [50, 55, 52], [45, 50, 48], [30, 40, 35], [55, 60, 58], [40, 45, 42]
])

# Метод локтя для выбора оптимального k
inertias = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# График метода локтя
plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), inertias, marker='o', linestyle='--')
plt.xlabel("Число кластеров (k)")
plt.ylabel('Inertia (Сумма квадратов расстояний)')
plt.title('Метод локтя для выбора оптимального k')
plt.grid()
plt.show()

# Кластеризация с оптимальным k=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(" ")

# 3D визуализация
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', s=60)
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2],
          marker='X', s=200, c='red', label='Центроиды')
ax.set_xlabel('Математика')
ax.set_ylabel('Физика')
ax.set_zlabel('Информатика')
plt.title('3D-кластеризация студентов по трём предметам')
plt.legend()
plt.show()

# Анализ результатов
df = pd.DataFrame(X, columns=['Математика', 'Физика', 'Информатика'])
df['Кластер'] = labels

print("Центроиды кластеров:")
for i, centroid in enumerate(centroids):
    print(f"Кластер {i}: {centroid.round(2)}")

print("\nСредние баллы по кластерам:")
print(df.groupby('Кластер').mean().round(2))

print("\nКоличество студентов в кластерах:")
print(df['Кластер'].value_counts())
```

# Задание 1. Самостоятельный пример.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Средний балл, пропуский лекций, посещение лекций
X = np.array([
    [85, 2, 3], [72, 5, 1], [90, 1, 5], [68, 8, 0], [95, 0, 6],
    [68, 8, 0], [95, 0, 6], [60, 10, 0], [78, 4, 2], [82, 3, 4],
    [55, 12, 0], [88, 1, 5], [65, 7, 1], [92, 0, 7]
])

# Метод локтя для выбора оптимального k
inertias = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# График метода локтя
plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), inertias, marker='o', linestyle='--')
plt.xlabel("Число кластеров (k)")
plt.ylabel('Inertia (Сумма квадратов расстояний)')
plt.title('Метод локтя для выбора оптимального k')
plt.grid()
plt.show()

# Кластеризация с оптимальным k=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(" ")

# 3D визуализация
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', s=60)
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2],
          marker='X', s=200, c='red', label='Центроиды')
ax.set_xlabel('Средний балл')
ax.set_ylabel('Пропуски лекций')
ax.set_zlabel('Посещение лекций')
plt.title('3D-кластеризация по трем параметрам')
plt.legend()
plt.show()

# Анализ результатов
df = pd.DataFrame(X, columns=['Средний балл', 'Пропуски лекций', 'Посещение лекций'])
df['Кластер'] = labels

print("Центроиды кластеров:")
for i, centroid in enumerate(centroids):
    print(f"Кластер {i}: {centroid.round(2)}")

print("\nСредние баллы по кластерам:")
print(df.groupby('Кластер').mean().round(2))

print("\nКоличество студентов в кластерах:")
print(df['Кластер'].value_counts())
```

```python
print("Анализ кластеров:")

for i, centroid in enumerate(centroids):
    avg_score = centroid[0]
    misses = centroid[1]
    attends = centroid[2]

    print(f"\nКластер {i}:")

    # Характеристика кластера
    if avg_score > 85 and misses < 2 and attends > 4:
        print("Отличники")
    elif avg_score > 70 and misses < 6 and attends > 2:
        print("Хорошисты")
    else:
        print("Группа риска")
```

# Задание 2. Самостоятельный пример.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

X = np.array([
    [18, 25, 92], [12, 45, 75], [20, 20, 98], [8, 60, 65], [15, 30, 85],
    [5, 90, 50], [17, 28, 90], [10, 50, 70], [19, 22, 95], [6, 75, 55],
    [14, 35, 80], [16, 30, 88]
])

inertias = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), inertias, marker='o', linestyle='--')
plt.xlabel("Число кластеров (k)")
plt.ylabel('Inertia (Сумма квадратов расстояний)')
plt.title('Метод локтя для выбора оптимального k')
plt.grid()
plt.show()

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(" ")

df = pd.DataFrame(X, columns=['Средний_балл', 'Пропуски_лекций', 'Посещение_лекций'])
df['Кластер'] = labels

plt.figure(figsize=(15, 12))

plt.subplot(2, 2, 1)
for cluster in sorted(df['Кластер'].unique()):
    cluster_data = df[df['Кластер'] == cluster]
    plt.scatter(cluster_data['Средний_балл'], cluster_data['Пропуски_лекций'],
               label=f'Кластер {cluster}', s=80, alpha=0.8)
plt.xlabel('Средний балл')
plt.ylabel('Пропуски лекций')
plt.title('Средний балл vs Пропуски лекций')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
for cluster in sorted(df['Кластер'].unique()):
    cluster_data = df[df['Кластер'] == cluster]
    plt.scatter(cluster_data['Средний_балл'], cluster_data['Посещение_лекций'],
               label=f'Кластер {cluster}', s=80, alpha=0.8)
plt.xlabel('Средний балл')
plt.ylabel('Посещение лекций')
plt.title('Средний балл vs Посещение лекций')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 3)
for cluster in sorted(df['Кластер'].unique()):
    cluster_data = df[df['Кластер'] == cluster]
    plt.scatter(cluster_data['Пропуски_лекций'], cluster_data['Посещение_лекций'],
               label=f'Кластер {cluster}', s=80, alpha=0.8)
plt.xlabel('Пропуски лекций')
plt.ylabel('Посещение лекций')
plt.title('Пропуски лекций vs Посещение лекций')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['blue', 'green', 'orange']
for i in range(3):
    cluster_data = df[df['Кластер'] == i]
    ax.scatter(cluster_data['Средний_балл'],
               cluster_data['Пропуски_лекций'],
               cluster_data['Посещение_лекций'],
               c=colors[i], label=f'Кластер {i}', s=80, alpha=0.8)

ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2],
          marker='X', s=200, c='red', label='Центроиды')
ax.set_xlabel('Средний балл')
ax.set_ylabel('Пропуски лекций')
ax.set_zlabel('Посещение лекций')
plt.title('3D-кластеризация по трем параметрам')
plt.legend()
plt.show()

print("Центроиды кластеров:")
centroids_df = pd.DataFrame(centroids,
                           columns=['Средний_балл', 'Пропуски_лекций', 'Посещение_лекций'])
print(centroids_df.round(2))

print("\nСредние значения по кластерам:")
cluster_stats = df.groupby('Кластер').mean().round(2)
print(cluster_stats)

print("\nКоличество студентов в кластерах:")
print(df['Кластер'].value_counts().sort_index())

print("")
print("ДЕТАЛЬНЫЙ АНАЛИЗ КЛАСТЕРОВ")

for cluster in sorted(df['Кластер'].unique()):
    cluster_data = df[df['Кластер'] == cluster]
    print(f"\n--- Кластер {cluster} ---")
    print(f"Количество студентов: {len(cluster_data)}")
    print(f"Средний балл: {cluster_data['Средний_балл'].mean():.1f}")
    print(f"Средние пропуски: {cluster_data['Пропуски_лекций'].mean():.1f}%")
    print(f"Среднее посещение: {cluster_data['Посещение_лекций'].mean():.1f}%")

    avg_score = cluster_data['Средний_балл'].mean()
    attendance = cluster_data['Посещение_лекций'].mean()

    if avg_score >= 16 and attendance >= 85:
        status = "Успешные студенты"
    elif avg_score >= 12 and attendance >= 70:
        status = "Студенты с риском"
    else:
        status = "Студенты высокого риска"

    print(f"Статус: {status}")
```
